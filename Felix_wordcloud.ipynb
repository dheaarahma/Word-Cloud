```python
# Read and convert an mask image. It should have a white (not transparent) background with a black object.
from PIL import Image 
import numpy as np 

mask_img = np.array(Image.open("./Felix159.png"))
wordcloud = WordCloud(
    width = 3000, 
    height = 2000,
    background_color="white", 
    max_font_size=300, min_font_size=10, max_words=300, mask=mask_img
).generate(text)
```

```python
# store to file
wordcloud.to_file("felix_birthday.png")
```

```python
# Show the image
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
